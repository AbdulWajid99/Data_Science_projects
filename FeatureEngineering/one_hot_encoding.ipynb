```python
import numpy as np
import pandas as pd


df = pd.read_csv("train.csv", usecols=['X1','X2','X3','X4','X5','X6'])
df.head()
```

```python
#One hot encoding
# here we are getting 123 columns out of 5 columns not Cool so we apply one hot ecoding for multi value
pd.get_dummies(df)
```

```python
pd.get_dummies(df, drop_first=True)
```

```python
#multi value: take top 10 categories form each features

df.X2.value_counts().head(20)
```

```python
# list with 10 most frequent variables

top_10 = [x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
# set all in top 10 to 1 and not in top 10 to 0 in this X2

for categories in top_10:
    df[categories] = np.where(df['X2']==categories,1,0)
    
df[['X2']+top_10].head(40)
```

```python

# now create dummy variables for each column
def one_hot_encoding_multi_value(dataframe,column_name,top_10_categories):
    for categories in top_10_categories:
        dataframe[column_name+'_'+categories] = np.where(dataframe[column_name]==categories,1,0)
    
one_hot_encoding_multi_value(df, 'X2' ,top_10)
df.head()
```

```python
#now for X1

top_10_x1 = [x for x in df.X1.value_counts().sort_values(ascending=False).head(10).index]

one_hot_encoding_multi_value(df, 'X1' ,top_10_x1)
df.head()
```

```python
#now for X3

top_10_x3 = [x for x in df.X3.value_counts().sort_values(ascending=False).head(10).index]

one_hot_encoding_multi_value(df, 'X3' ,top_10_x3)
df.head()
```

```python
#now for X4

top_10_x4 = [x for x in df.X4.value_counts().sort_values(ascending=False).head(10).index]

one_hot_encoding_multi_value(df, 'X4' ,top_10_x4)
df.head()
```
